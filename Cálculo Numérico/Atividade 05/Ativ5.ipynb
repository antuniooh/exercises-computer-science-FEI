## EXERCÍCIO 1

```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
X = [1,2,3,4]
X = np.array(X)
y = [2.5,3.83,4.96,6.00]
X = X.reshape(-1, 1)
modelo = LinearRegression().fit(X, y)
print(modelo.intercept_)
print (modelo.coef_)
plt.scatter(X, y)
plt.plot(X, modelo.predict(X), color = 'red')
modelo._residues
```

## Exercício 2

```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
V= [2,3,4,5,7,10]
V = np.array(V)
i = [5.2,7.8,10.7,13,19.3,27.5]
V = V.reshape(-1, 1)
modelo = LinearRegression().fit(V, i)
print(modelo.intercept_)
print (modelo.coef_)
plt.scatter(V, i)
plt.plot(V, modelo.predict(V), color = 'red')
# valor de i para V=3.5
yr=modelo.intercept_ + modelo.coef_ * 3.5
print(yr)
#valor de V para i =15
v=sy.Symbol('v')
i=modelo.intercept_ + modelo.coef_ * v -15
print(sy.solve(i))
modelo._residues
```
